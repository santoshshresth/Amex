```python
import pandas as pd
df=pd.read_csv("train_data.csv")
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df = df.sort_values(['customer_ID', 'S_2'])
df
```

```python
##Create Utilization (Ratio) Features
df['util_1'] = df['B_1'] / (df['B_2'] + 1e-5)
df['util_2'] = df['D_39'] / (df['P_2'] + 1e-5)
df['util_3'] = df['B_3'] / (df['B_2'] + 1e-5)
```

```python
##Creating  capacity features
df['capacity_left_1'] = df['B_2'] - df['B_1']
df['capacity_left_2'] = df['P_2'] - df['D_39']
```

```python
##Adding behaviour and customer level
df['util_1_diff'] = df.groupby('customer_ID')['util_1'].diff()
```

```python
final_df = df.groupby('customer_ID').agg({
    'util_1': ['mean', 'max'],
    'util_2': ['mean'],
    'capacity_left_1': ['mean'],
    'util_1_diff': ['mean']
})
```

```python
final_df = final_df.fillna(0)
final_df
```
